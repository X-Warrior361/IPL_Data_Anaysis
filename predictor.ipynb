```python
import sqlalchemy
import pandas as pd
import numpy as np
```

```python
engine = 'mysql+mysqldb://root:password@localhost:3306/ipl'   # Change according to your local database
conn = sqlalchemy.create_engine(engine)
```

```python
sql = sqlalchemy.text("select * from all_matches")
all_matches = pd.read_sql_query(sql, conn.connect())
all_matches.head()
```

```python
sql = sqlalchemy.text("select * from match_details")
match_data = pd.read_sql_query(sql, conn.connect())
match_data.head()
```
